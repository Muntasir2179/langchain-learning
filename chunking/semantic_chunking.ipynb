## Semantic Splitter

```python
from rich import print
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_experimental.text_splitter import SemanticChunker
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
loader = PyMuPDFLoader("docs/The-Army-Regulations.pdf")
data = loader.load()

print(type(data))
```

```python
print(data[:3])
```

```python
# loads the pdf in each page
len(data)
```

```python
# printing a single page
print(data[30].metadata)
```

```python
print(data[30].page_content)
```

```python
embedding = HuggingFaceEmbeddings(model_name="mixedbread-ai/mxbai-embed-large-v1",
                                  cache_folder="../embedding_model",
                                  model_kwargs={"device": 'cuda:0'})
```

```python
# defining the text splitter
text_splitter = SemanticChunker(embeddings=embedding, 
                                breakpoint_threshold_type='percentile')

# splitting documents to create chunks
documents = text_splitter.split_documents(documents=data)
```

```python
print(len(documents))
```

```python
print(documents[120])
```

```python
print(documents[600])
```

```python
doc_lengths = [len(doc.page_content) for doc in documents]
```

```python
len(doc_lengths)
```

## Processing the documents

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 6))

# Plot the list
plt.plot(doc_lengths, marker='o')

# Adding labels and title
plt.title('Lengths of contexts')
plt.xlabel("i'th context")
plt.ylabel('Length')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(20, 6))

# Plot the list
plt.plot([item for item in doc_lengths if item < 150], marker='o')

# Adding labels and title
plt.title('Lengths of contexts')
plt.xlabel("i'th context")
plt.ylabel('Length')

# Show the plot
plt.show()
```

## Creating vector database

```python
import os
from langchain_community.vectorstores import Chroma

# setting up vector db directory
persist_directory = os.path.join(os.path.dirname(os.path.abspath("__file__")), "semantic_chunk_db")

db = Chroma.from_documents(
    documents=documents,
    embedding=embedding,
    persist_directory=persist_directory
)
```
